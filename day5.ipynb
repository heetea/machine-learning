# 과제 1번

```python
class Set:
    def __init__(self, member=[]):
        self.member = member
        self.result =  {i for i in self.member}
            
    def append(self, a):
        if a in member:
            return self.result
        else:
            temp = member.append(a)
            return {i for i in temp}
    def delete(self, a):
        if a in self.result:
            self.temp = member.remove(a)
            self.dele = {i for i in self.temp}
            return self.dele
        else:
            return "없어요!"
           
    def union(self, s2):
        self.mem = self.member + s2.member 
        self.union = {i for i in self.mem}
        return self.union
    
    def difference(self, s2):
        self.dif = []
        for i in self.mem:
            if i not in s2.member:
                self.dif.append(i)
            self.difference = {i for i in self.dif}                
        return self.difference    

    def intersection(self, s2):
        self.inter = []
        for i in self.member:
            if i in s2.member:
                self.inter.append(i)
            self.intersection = {i for i in self.inter} 
        return self.intersection    


    def __add__(self, other):
        return self.union
       
    def __sub__(self, other):
        return self.difference
       
    def __truediv__(self,other):
        return self.intersection
```

```python
a = Set([1, 2, 3])
b = Set([2, 3, 4])
c = a.union(b)
print(c)

d = a.difference(b)

print(d)

e = a.intersection(b)

print(e)
c = a + b
print(c)
d = a - b
print(d)
e = a / b
print(e)
```

# 과제 2번

```python
class person:
    def __init__(self, name, age, department):
        self.name = name
        self.age = age
        self.department = department
    def get_name(self):
        return self.name
    
class Student(person):
    def __init__(self, name, age, department, s_id, GPA ):
        person.__init__(self, name, age, department)
        self.s_id = s_id
        self.GPA = GPA
    def reg_advisor(self, prof):
        self.advisor = prof
    def print_info(self):
        print("제 이름은", self.name, ", 나이는", self.age, ", 학과는", self.department, ",지도교수님은",self.advisor.name, "입니다")

class Professor(person):
    def __init__(self, name, age, department, position, laboratory):
        person.__init__(self, name, age, department)
        self.position = position
        self.laboratory = laboratory
        self.student = []
    def reg_student(self, student):
        self.student.append(student)
    def print_info(self):
        print("제 이름은", self.name, ", 나이는", self.age, ", 학과는", self.department, ",지도학생은", end=" ")
        for i in self.student:
            print(i.name, end=", ")
        print("입니다")
        
        
        
```

```python
stu1 = Student('Kim', 30, 'Computer', 20001234, 4.5)
stu2 = Student('Lee', 22, 'Computer',20101234, 0.5)
prof1 = Professor('Lee', 55, 'Computer','Full', 'KLE')

stu1.reg_advisor(prof1)

stu2.reg_advisor(prof1)
prof1.reg_student(stu1)
prof1.reg_student(stu2)
stu1.print_info()
stu2.print_info()
prof1.print_info()
```
